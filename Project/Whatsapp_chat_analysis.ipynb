```python
import re
import pandas as pd
```

```python
f = open('/content/what.txt','r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}'
```

```python
pattern= '\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\s:\s-'
```

```python
messages = re.split(pattern, data)
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame.from_dict({'user_message': messages, 'message_date': dates})

df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M %p - ')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```
